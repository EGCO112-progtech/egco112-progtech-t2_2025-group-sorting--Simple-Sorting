# Using SKLEARN

```python
import sklearn.preprocessing as preprocessing
scaler = preprocessing.StandardScaler()

scaler.fit(df.iloc[:,1:5])
```

```python
print(scaler.var_)
```

```python
#result in numpy array
X_std=scaler.transform(df.iloc[:,1:5])
X_std
```

```python
pd_std=pd.DataFrame(X_std, columns=df.iloc[:,1:5].columns)
```

```python
pd_std.describe()
```

```python
pd_std['Species']=df['Species']
```

```python
pd_std.describe()
```
